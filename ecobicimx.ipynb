```python
# concatenar Hora_Retiro y Fecha_Retiro
viajes['fecha_hora_retiro'] = viajes.Fecha_Retiro + ' ' + viajes.Hora_Retiro

# cambiar de str a datetime
viajes['fecha_hora'] = pd.to_datetime(viajes['fecha_hora_retiro'], format= '%d/%m/%Y %H:%M:%S' )


# reindexar el dataframe
viajes.index = viajes.fecha_hora
```

```python
# limpiar valores de otros años
viajes = viajes.loc['2018-10']
viajes
```

```python
# resample y agregacion por dia de mes
viajes_resample_day = viajes.Bici.resample('h').count()
viajes_resample_day
```

```python
# asignar día de la semana
df_resample = pd.concat([viajes_resample_day], axis=1) # axis = 1: indica que la serie se fusionará como una columna
df_resample
```

```python
df_resample['dayofweek'] = df_resample.index.dayofweek #toma el día de la semana donde  0 es lunes del indice, creando 
#una nueva columna 
df_resample
```

```python
# solo tomar los días de lunes a viernes
df_mon_to_fri = df_resample[df_resample.dayofweek.isin([0,1,2,3,4])].Bici #método isin filtra los vales 
# .sin([valor]), en este caso los días de la semana son lunes-viernes, 5=sabado, 6=domingo
#df_mon_to_fri
```
